Em relação ao vírus, a Covid-19 é conhecida como novo coronavírus *(ou Sars-CoV-2)* por fazer parte de uma família maior de vírus já descoberta por cientistas, tais como o MERS-CoV e SARS-CoV. De fácil pronúncia e sem relação com nenhum país, o vírus mais conhecido do mundo tem a junção das palavras *(em inglês)* *corona(co)*, *virus(vi)*, *disease(d)* e *19*, referente ao ano do seu surgimento, em 2019. 

Uma das preocupações com o novo coronavírus é que se espalha facilmente pelo ar nas gotículas liberadas pela fala, tosse ou espirro, alcançando assim um grande número de pessoas. O contato com o vírus também ocorre de maneira direta, com abraços e beijos, ou indireta, quando as mãos contaminadas tocam a boca, o nariz ou os olhos. 

Em contato com as vias respiratórias, o vírus entra em fase de incubação e, a partir de então, começa a se espalhar pelo corpo e causar as primeiras reações. Na maior parte dos casos, o corpo combate de forma eficiente, causando pouco ou nenhum sintoma em até 14 dias após a exposição ao vírus. Vale ressaltar aqui que uma pessoa pré-sintomática também transmite o vírus, viu? 

Os sintomas mais comuns causados pelo vírus, na maioria dos casos, é febre, tosse e dificuldade para respirar. Esses sinais se assemelham muito a um resfriado comum, de nível leve a moderado, e pode facilmente ser confundido nos estágios iniciais da doença. Em outros casos, podem surgir sintomas como fadiga, perda de olfato/paladar, náusea, etc. Nos casos mais grave, o vírus causa lesões nos alvéolos pulmonares e dá início ao processo inflamatório. Esses alvéolos são pequenas bolsas responsáveis pela troca gasosa entre o pulmão e corrente sanguínea. Nessa fase da doença, sintomas como falta de ar e baixa oxigenação dos órgãos são comuns, e por isso necessitam de atenção especial da parte médica. Além disso, já se sabe que a infecção causada pelo novo coronavírus pode também atingir órgãos como fígado, coração e rins.

### **Dicionário de Variáveis**

Como visto anteriormente, o dataset possui 113406 valores de entrada para países e datas específicas com 62 variáveis informativas e relacionadas aos casos do novo coronavírus. Mas só essa informação não nos diz muito, uma vez que é necessário conhecer o que cada uma delas significa antes de começar a transformar os dados brutos em informação com valor agregado. Por esse motivo, foi construído um dicionário de variáveis a partir da documentação original e que pode ser acessada [aqui](https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-codebook.csv).

**Casos confirmados**
* `total_cases` - Total de casos.
* `new_cases` - Novos casos.
* `new_cases_smoothed` - Novos casos confirmados e suavizado por sete dias.
* `total_cases_per_million` - Total de casos *(por milhão de pessoas)*.	
* `new_cases_per_million` - Novos casos *(por milhão de pessoas)*.
* `new_cases_smoothed_per_million` - Novos casos confirmados e suavizado por sete dias *(por milhão de pessoas)*.

**Mortes confirmadas**
* `total_deaths` - Total de mortes.
* `new_deaths` - Novas mortes.
* `new_deaths_smoothed` -	Novas mortes confirmadas e suavizadas por sete dias.
* `total_deaths_per_million` - Total de mortes *(por milhão de pessoas)*.
* `new_deaths_per_million` - Novas mortes *(por milhão de pessoas)*.
* `new_deaths_smoothed_per_million` - Novas mortes confirmadas e suavizadas por sete dias *(por milhão de pessoas)*.

**Hospital e Unidade de Terapia Intensiva (UTI)**
* `icu_patients` - Número de pacientes em UTI no dia específico.
* `icu_patients_per_million` - Número de pacientes em UTI no dia específico *(por milhão de pessoas)*.
* `hosp_patients` - Número de pacientes hospitalizados no dia específico.
* `hosp_patients_per_million` - Número de pacientes hospitalizados no dia específico *(por milhão de pessoas)*.
* `weekly_icu_admissions` - Número no pacientes admitidos em UTI na semana específica.
* `weekly_icu_admissions_per_million` - Número no pacientes admitidos em UTI na semana específica *(por milhão de pessoas)*.
* `weekly_hosp_admissions` - Número no pacientes admitidos em hospitais na semana específica.
* `weekly_hosp_admissions_per_million` - Número no pacientes admitidos em hospitais na semana específica *(por milhão de pessoas)*.

**Resposta política**
* `stringency_index` - Índice de restrição dos governos diante da pandemia de covid-19, onde **0** indica **sem restrições** e **100** indica **restrição mais alta**.

**Taxa de reprodução**
* `reproduction_rate` - Estimativa em tempo real da taxa de reprodução da covid-19. Para mais informações, consulte o artigo de [Arroyo-Marioli e colaboradores (2021)](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0244474).

**Testes e positividades**
* `total_tests` - Total de testes aplicados.
* `new_tests` - Novos testes aplicados *(calculados no dia consecutivo)*.
* `total_tests_per_thousand` - Total de testes *(por mil pessoas)*.
* `new_tests_per_thousand` - Novos testes *(por mil pessoas)*.
* `new_tests_smoothed` - Novos testes aplicados e suavizados por sete dias.
* `new_tests_smoothed_per_thousand` - Novos testes aplicados e suavizados por sete dias *(por mil de pessoas)*.
* `positive_rate` - Testes positivos, dada como uma média contínua de sete dias *(inverso de `test_per_case`)*.
* `tests_per_case` - Testes com novos casos confirmados, dado como uma média contínua de sete dias *(inverso de `positive_rate`)*.
* `tests_units` - Unidades usadas para relatar os dados de teste.

**Vacinação**
* `total_vaccinations` - Total de doses administradas.
* `people_vaccinated` - Total de pessoas que receberam ao menos uma dose.
* `people_fully_vaccinated` - Total de pessoas que receberam o total das doses indicada nos protocolos.
* `total_boosters` - Número de doses de reforço além do número prescrito pelo protocolo de vacinação. 
* `new_vaccinations` - Novas doses administradas *(calculadas no dia consecutivo)*.
* `new_vaccinations_smoothed` - Novas doses administradas *(calculadas no dia consecutivo)* e suavizados por sete dias.
* `total_vaccinations_per_hundred` - Total de doses administradas *(por 100 pessoas)*.
* `people_vaccinated_per_hundred` - Total de pessoas que receberam ao menos uma dose *(por 100 pessoas)*.
* `people_fully_vaccinated_per_hundred` - Total de pessoas que receberam todas as doses indicada nos protocolos *(por 100 pessoas)*.
* `total_boosters_per_hundred` - Número de doses de reforço além do número prescrito pelo protocolo de vacinação *(por 100 pessoas)*.
* `new_vaccinations_smoothed_per_million` - Novas doses administradas *(calculadas no dia consecutivo)* e suavizados por sete dias* (por milhão de pessoas)*.

**Outros**
* `iso_code` - Código identificador do paíse com três letras *(ISO 3166-1 alpha-3)*.
* `continent` - Continente da localização geográfica.
* `location` - Localização geográfica.
* `date` - Data da observação.
* `population` - População no ano de 2020.
* `population_density` - Densidade populacional, calculado a partir da divisão do número de pessoas dividido por área terrestre *(em quilômetros quadrados)*, no ano mais recentes.
* `median_age` - Idade mediana da população, feita pela projeção da *Organização das Nações Unidas (ONU)* para 2020.
* `aged_65_older` - População com 65 anos ou mais no ano mais recente.
* `aged_70_older` - População com 70 anos ou mais em 2015.
* `gdp_per_capita` - Produto Interno Bruto *(PIB)* em paridade de poder de compra (dolar internacional constante de 2021), no ano mais recente.
* `extreme_poverty` - Parcela da população em extrema pobreza, no ano mais recente disponível desde 2010.
* `cardiovasc_death_rate` - Índice de mortalidade por doença cardiovascular em 2017 *(valor anual de mortes por cem mil pessoas)*.
* `diabetes_prevalence` - Prevalência de diabetes nas pessoas entre 19 e 79 anos em 2017.
* `female_smokers` - Pessoas do sexo feminino que fumam, no ano mais recente.
* `male_smokers` - Pessoas do sexo masculo que fumam, no ano mais recente.
* `handwashing_facilities` - Pessoas com instalações básicas para lavagem das mãos, no ano mais recente.
* `hospital_beds_per_thousand` - Camas de hospital por mil pessoas, no ano mais recente, desde 2010.
* `life_expectancy` - Expectativa de vida em 2019.
* `human_development_index` - Índice composto por três dimensões, sendo elas *vida longa e saudável*, *conhecimento* e *padrão de vida decente*. Os valores do Índice de Desenvolvimento Humano de 2019 podem ser consultados [aqui](http://hdr.undp.org/en/indicators/137506).
* `excess_mortality` - Excesso de mortalidade *(P-Valor)* para todas as idades. Para mais informações sobre definição e cálculo, consulte a documentação [aqui](https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality).

Para a análise das variáveis `new_cases` e `total_cases`, foram selecionadas os 7 países com mais casos no dia da coleta dos dados. Na análise da variável `new_cases`, os Estados Unidos aparece em disparada em relação aos demais países. Mesmo sendo um dos primeiros países a começar com a vacinação da população, fatores como a variante Delta, adultos não vacinados e queda na efetividade da vacinação deixam as autoridades em alerta para uma [terceira onda de covid-19](https://bit.ly/2YvmL6R) no território. Inclusive, a alta de casos confirmados diariamente incide diretamente no total de casos de cada país. No caso da variável `total_cases`, o Estados Unidos continua em primeiro lugar com o maior número, seguido de países como Índia e Brasil. 




```
#@title
# new_cases + total_cases
top7_new_cases = df.loc[df.date == '2021-08-31', ['location', 'new_cases']].sort_values(by='new_cases', ascending=False)[0:7]
top7_total_cases = df.loc[df.date == '2021-08-31', ['location', 'total_cases']].sort_values(by='total_cases', ascending=False)[0:7]

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))
top7_new_cases.plot(x='location', y='new_cases', kind='bar', ax=ax[0], color='grey', alpha=.4)
ax[0].set_title('Novos casos de COVID-19 no mundo', loc='left', fontsize=14, fontweight='bold')
ax[0].text(-.5, top7_new_cases.new_cases.max()+40000, 'Número de casos confirmados para o dia 31 de agosto de 2021.', fontsize=13)

ax[0].set_xlabel('')
ax[0].set_xticklabels(['Estados Unidos', 'Índia', 'Reino Unido', 'Irã', 'Brasil', 'Turquia',' Malásia'], rotation=360, fontsize=12)

ax[0].set_ylim(0, top7_new_cases.new_cases.max()+50000)
ax[0].get_yaxis().set_ticks([])

for i, v in enumerate(top7_new_cases.new_cases):
  ax[0].text(i, v+3000, str(int(v)), horizontalalignment='center', fontsize=12)

ax[0].get_legend().remove()

top7_total_cases.plot(x='location', y='total_cases', kind='bar', ax=ax[1], color='grey', alpha=.4)
ax[1].set_title('Total de casos de COVID-19 no mundo', loc='left', fontsize=14, fontweight='bold')
ax[1].text(-.5, top7_total_cases.total_cases.max()+13000000, 'Número de casos confirmados para o dia 31 de agosto de 2021.', fontsize=13)

ax[1].set_xlabel('')
ax[1].set_xticklabels(['Estados Unidos', 'Índia', 'Brasil', 'França', 'Reino Unido', 'Rússia', 'Turquia'], rotation=360, fontsize=12)

ax[1].set_ylim(1, top7_total_cases.total_cases.max()+16000000)
ax[1].get_yaxis().set_ticks([])

ax[1].get_legend().remove()

for i, v in enumerate(top7_total_cases.total_cases):
  ax[1].text(i, v+1000000, str(int(v)), horizontalalignment='center', fontsize=12)

for i in ax:
  i.spines['top'].set_visible(False)
  i.spines['left'].set_visible(False)
  i.spines['right'].set_visible(False)

plt.tight_layout()
```

Para a análise das variáveis `new_deaths` e `total_deaths`, foram selecionadas  os 7 países com mais casos no dia da coleta dos dados. Os Estados Unidos segue com o maior número de mortes para o dia, segundo a variável `new_deaths`. Segundo aponta a [CNN Brasil](https://bit.ly/2VpXn1c), os Estados Unidos poderiam ter sido evitadas 40% das mortes se o atual governo não tivesse rejeitado publicamente as ameaças da covid-19, inclusive, minimizando qualquer ação que pudesse combater o vírus. Com o mesmo comportamento, o [negacionismo na pandemia](https://bit.ly/3yYaynP) levou o Brasil ao segundo lugar no número de mortes para o dia analisado. Como consequência da má gestão, Estados Unidos e Brasil permanecem em primeiro e segundo lugar, respectivamente, para a variável `total_deaths`.


```
#@title
# new_deaths + total_deaths
top7_new_deaths = df.loc[df.date == '2021-08-31', ['location', 'new_deaths']].sort_values(by='new_deaths', ascending=False)[0:7]
top7_total_deaths = df.loc[df.date == '2021-08-31', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[0:7]

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

top7_new_deaths.plot(x='location', y='new_deaths', kind='bar', ax=ax[0], color='grey', alpha=.4)
ax[0].set_title('Novas mortes por COVID-19 no mundo', loc='left', fontsize=14, fontweight='bold')
ax[0].text(-.5, top7_new_deaths.new_deaths.max()+500, 'Número de casos confirmados para o dia 31 de agosto de 2021.', fontsize=13)

ax[0].set_xlabel('')
ax[0].set_xticklabels(['Estados Unidos', 'Brasil', 'México', 'Rússia', 'Irã', 'Indonésia', 'Índia'], rotation=360, fontsize=12)

ax[0].set_ylim(0, top7_new_deaths.new_deaths.max()+600)
ax[0].get_yaxis().set_ticks([])

for i, v in enumerate(top7_new_deaths.new_deaths):
  ax[0].text(i, v+50, str(int(v)), horizontalalignment='center', fontsize=12)

ax[0].get_legend().remove()

top7_total_deaths.plot(x='location', y='total_deaths', kind='bar', ax=ax[1], color='grey', alpha=.4)
ax[1].set_title('Total de mortes por COVID-19 no mundo', loc='left', fontsize=14, fontweight='bold')
ax[1].text(-.5, top7_total_deaths.total_deaths.max()+200000, 'Número de casos confirmados para o dia 31 de agosto de 2021.', fontsize=13)

ax[1].set_xlabel('')
ax[1].set_xticklabels(['Estados Unidos', 'Brasil', 'Índia', 'México', 'Peru', 'Rússia', 'Indonésia'], rotation=360, fontsize=12)

ax[1].set_ylim(1, top7_total_deaths.total_deaths.max()+250000)
ax[1].get_yaxis().set_ticks([])

ax[1].get_legend().remove()

for i, v in enumerate(top7_total_deaths.total_deaths):
  ax[1].text(i, v+20000, str(int(v)), horizontalalignment='center', fontsize=12)

for i in ax:
  i.spines['top'].set_visible(False)
  i.spines['left'].set_visible(False)
  i.spines['right'].set_visible(False)

plt.tight_layout() 
```

Em relação as vacinas, comparar as variávels `total_vaccinations` e `people_fully_vaccinated` em números absolutos não seria justo. Espera-se países com maior população apliquem mais doses, e essa relação não indica necessariamente uma melhor gestão contra a covid-19. Sendo assim, foram analisadas as variáveis `total_vaccinations_per_hundred` e `people_fully_vaccinated_per_hundred`, onde são verificadas o número de doses administradas a cada 100 pessoas. Para ambas as variáveis foram selecionadas os 7 países mais casos no dia da coleta dos dados.

Para a variável `total_vaccinations_per_hundred`, os dados mostram pouca diferença entre os países. O país que mais administrou doses no dia foi o Emirados Árabes Unidos, seguido do Uruguai e Israel.  

Independente da quantidade de doses administradas, é a partir da segunda dose da vacina que o corpo passa a produzir anticorpos suficientes para combater a covid-19, garantindo a eficácia obtida em testes clínicos. Vale lembrar que existem vacinas de dose única também. Sendo assim, para a variável `people_fully_vaccinated_per_hundred`, a Islândia, Ilhas Cayman e Portugal são exemplos para o mundo. 

```
#@title
# 'total_vaccinations_per_hundred' + 'people_fully_vaccinated_per_hundred'
top7_total_vaccinations_100 = df.loc[df.date == '2021-08-31', 
                                     ['location', 'total_vaccinations_per_hundred']
                                     ].sort_values(by='total_vaccinations_per_hundred', ascending=False)[0:7]
top7_people_fully_vaccinated_100 = df.loc[df.date == '2021-08-31', 
                                          ['location', 'people_fully_vaccinated_per_hundred']
                                          ].sort_values(by='people_fully_vaccinated_per_hundred', ascending=False)[0:7]


fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(9, 6))

top7_total_vaccinations_100.plot(x='location', y='total_vaccinations_per_hundred', kind='bar', ax=ax[0], color='grey', alpha=.4)
ax[0].set_title('Total de doses administradas contra a Covid-19 (por 100 pessoas)', loc='left', fontsize=14, fontweight='bold')
ax[0].text(-.5, top7_total_vaccinations_100.total_vaccinations_per_hundred.max()+40, 'Total de doses aplicadas até o dia 31 de agosto de 2021.', fontsize=13)

ax[0].set_xlabel('')
ax[0].set_xticklabels(['Emirados \nÁrabes Unidos', 'Uruguai', 'Israel', 'Islândia', 'Catar',  'Ilhas Caymans', 'Chile'], rotation=360, fontsize=12)

ax[0].set_ylim(0, top7_total_vaccinations_100.total_vaccinations_per_hundred.max()+50)
ax[0].get_yaxis().set_ticks([])

for i, v in enumerate(top7_total_vaccinations_100.total_vaccinations_per_hundred):
  ax[0].text(i, v+4, str(int(v)), horizontalalignment='center', fontsize=12)

ax[0].get_legend().remove()


top7_people_fully_vaccinated_100.plot(x='location', y='people_fully_vaccinated_per_hundred', kind='bar', ax=ax[1], color='grey', alpha=.4)
ax[1].set_title('Total de de pessoas vacinadas contra a COVID-19 (por 100 pessoas)', loc='left', fontsize=14, fontweight='bold')
ax[1].text(-.5, top7_people_fully_vaccinated_100.people_fully_vaccinated_per_hundred.max()+15, 'Total de pessoas vacinadas até o dia 31 de agosto de 2021.', fontsize=13)

ax[1].set_xlabel('')
ax[1].set_xticklabels(['Islândia', 'Ilhas Caymans', 'Portugal', 'Ilha de Man', 'Catar', 'Dinamarca', 'Uruguai'], rotation=360, fontsize=12)

ax[1].set_ylim(1, top7_people_fully_vaccinated_100.people_fully_vaccinated_per_hundred.max()+20)
ax[1].get_yaxis().set_ticks([])

ax[1].get_legend().remove()

for i, v in enumerate(top7_people_fully_vaccinated_100.people_fully_vaccinated_per_hundred):
  ax[1].text(i, v+2, str(int(v)), horizontalalignment='center', fontsize=12)

for i in ax:
  i.spines['top'].set_visible(False)
  i.spines['left'].set_visible(False)
  i.spines['right'].set_visible(False)

plt.tight_layout() 
```

## **E o Brasil?**

Para avaliar o desempenho de cada país durante a pandemia, o Instituto Lowy, de Sidney/Australia, verificou quais impactos que a geografia, os sistemas políticos, o tamanho da população e o desenvolvimento econômico de cada nação tiveram na luta contra o novo coronavírus. O trabalho intitulado [Covid Performance Index](https://bit.ly/3k0tJco) *(ou Índice de Desempenho Covid)*, usou os casos confirmados, número de mortos, casos confirmados e de mortos por milhão, casos confirmados como proporção de teste e testes por mil pessoas para estabelecer notas de desempenho para cada país. Nesse sentido, os países são incluídos em um ranking com notas que vão de 0 *(pior desempenho)* a 100 *(melhor desempenho)*. Por questões de metodologia, países com menos de 100 casos confirmados ou com dados incompletos não fizeram parte da análise. 

O Ranking mais atual é do dia 13/03/2021, onde o Butão está em primeiro lugar no ranking *(nota 93)* e o Perú em centésimo quarto *(nota 4.8)*. O Brasil não aparece no raking mais atual, mas sim no anterior, do dia 09/01/2021, em último lugar *(nova 4.3)*. 

Tendo em vista a má gestão por parte dos governos para o enfrentamento contra a covid-19, vamos analisar os registros do Brasil.

```
# cria uma cópia com os dados do Brasil
df_br = df.loc[df.location == 'Brazil'].copy()

print(f'O primeiro registro de covid-19 no mundo foi dia {datetime.strftime(df.date.min(), "%d/%m/%Y")}, enquanto no Brasil foi no dia {datetime.strftime(df_br.date.min(), "%d/%m/%Y")}.')
```

Ao observar a variável `new_death`, é impossível não notar duas ondas. Sim, essas duas ondulações que se formam nos gráficos são chamadas de ondas, e são elas que preocupam a todos. Não existe um conceito bem definido sobre as ondas e o local onde elas começam ou terminam, mas é fato que isso representa um aumento de casos seguido de diminuição progressiva.

No caso do Brasil, especialistas acreditam que a primeira onda veio com a dispersão do vírus no país e que, após algumas semanas no ápice da contaminação, houve *'sucesso'* na implementação das medidas de isolamento, ocasionando a diminuição no número de casos.

E quando falo em *'sucesso'* é entre aspas mesmo, uma vez que alguns especialistas acreditam que nunca houve duas ondas, mas apenas uma. A má gestão da pandemia durante a primeira onda teve como consequência a segunda, com muito mais casos de covid-19. Aliás, não podemos deixar de falar que o surgimento de novas variantes, baixo ritmo de vacinação e flexibilização das restrições agravam ainda mais os problemas. Inclusive, especialistas afirmam que a situação pode ser ainda pior e que podemos estar [diante da terceira onda](https://bit.ly/3DZCn2K).

Analisando o gráfico da variável `new_deaths`, tivemos 839 vidas perdidas no dia 31 de agosto de 2021, valor abaixo da média. Apesar disso, nota-se um longo período na segunda onda com mais de mil mortes diárias. Apesar dos números em queda, não podemos deixar de citar que mais de seiscentos mil brasileiros já perderam suas vidas para a covid-19...

```
#@title
fig, ax = plt.subplots(figsize=(12,5))

ax.set_title('Novas mortes por Covid-19 no Brasil', loc='left', fontsize=14, fontweight='bold')
plt.text(pd.Timestamp('2020-02-01'), 4300, 'Números de óbitos entre os dias 26/02/2020 e 31/08/2021.', fontsize=13)
plt.bar('date', 'new_deaths', data=df_br, color='gray')
plt.plot('date', 'new_deaths_smoothed', data=df_br, color='red')

ax.set_ylabel('Número de casos', fontsize=12)
ax.set_xlabel('')

ax.axhline(y=df_br.new_deaths.mean(), linestyle='--', color='grey', linewidth=1)
ax.text(pd.Timestamp('2020-02-01'), df_br.new_deaths.mean()+70, f"Média de {round(df_br.new_deaths.mean(),2)} registros", color='grey', fontsize=10)
ax.text(pd.Timestamp('2021-08-31'), df_br[df_br.date == '2021-08-31'].new_deaths, f"{int(df_br.loc[14918].new_deaths)} óbitos", color='grey', fontsize=11)

for ax, spine in ax.spines.items():
  spine.set_visible(False)

plt.tight_layout()
```

Para a variável `new_cases`, também houve redução expressiva no número de contaminados, com 24589 novos casos no dia 31 de agosto de 2021. O comportamento da variável `new_cases` é semelhante ao gráfico anterior, mas com um longo período de novos casos diários acima da média.

```
#@title

fig, ax = plt.subplots(figsize=(12,5))

ax.set_title('Novos casos de Covid-19 no Brasil', loc='left', fontsize=14, fontweight='bold')
plt.text(pd.Timestamp('2020-01-30'), 130000, 'Números de novos casos entre os dias 26/02/2020 e 31/08/2021.', fontsize=13)
plt.bar('date', 'new_cases', data=df_br, color='gray')
plt.plot('date', 'new_cases_smoothed', data=df_br, color='red')

ax.set_ylim(0, df_br.new_cases.max()+18000)
ax.set_ylabel('Número de casos', fontsize=12)
ax.set_xlabel('')


ax.axhline(y=df_br.new_cases.mean(), linestyle='--', color='grey', linewidth=1)
ax.text(pd.Timestamp('2020-02-01'), df_br.new_cases.mean()+1500, f"Média de {round(df_br.new_cases.mean(),2)} registros", color='grey', fontsize=10)
ax.text(pd.Timestamp('2021-08-31'), df_br[df_br.date == '2021-08-31'].new_cases, f"{int(df_br.loc[14918].new_cases)} novos casos", color='grey', fontsize=11)

for ax, spine in ax.spines.items():
  spine.set_visible(False)

plt.tight_layout()
```

## **Conclusão**

A partir dos dados públicos disponibilizados pelo *Our World in Date*, foi possível verificar a situação da covid-19 no Brasil e no mundo. Para conhecer bem os dados, uma das primeiras ações foi a elaboração de um dicionário de variáveis. Também foi necessário fazer algumas adaptações antes de começar as análises, tais como modificar a variável `datetime64`, verificar valores ausentes e apagar registros desnecessários.

Foi possível verificar que países como Estados Unidos, Brasil e Índia foram os piores no controle da covid-19, permanecendo no topo de novos casos e total de casos confirmados para o dia 31 de agosto de 2021. Como consequência, os três países também aparecem com o maior número de pessoas mortas pelo novo coronavírus. Enquanto isso, países como Emirados Árabes Unidos, Uruguai e Israel foram os que mais administraram dose no dia em questão, enquanto Islândia, Ilhas Cayman e Portugal obtiveram as melhores posições em total de pessoas vacinadas. Ah, vale lembrar que os registros de vacinação em questão são para cada 100 pessoas, viu?

A pandemia da covid-19 também piorou a situação das populações mais vulneráveis. Os dados mostraram que os países do continente Africano com menos acesso a condições básicas de higiene são os mesmos com maior taxa de extrema pobreza e menor expectativa de vida. 

No caso do Brasil, a má gestão aliada a desinformação conseguiu transformar a pandemia em um verdadeiro massacre. O país teve o seu primeiro registro quase dois meses após o primeiro caso no mundo e, mesmo assim, conseguiu atingir a marca de mais de seiscentos mil mortos. Apesar dos números em queda e abaixo da média, ainda são centenas de vidas que se perdem diariamente. Esse mesmo comportamento se repete nos casos confirmados de COVID-19. 

Por fim, a história ainda não acabou. Com as novas variantes aparecendo e sem uma projeção para o fim da pandemia, seguimos com os cuidados básicos de proteção e rezando sempre por dias melhores. Amanhã sempre será um novo dia!
